Q.1. The marks awarded for an assignment set for a Year 8 class of 20 students were as
follows:
6 7 5 7 7 8 7 6 9 7 4 10 6 8 8 9 5 6 4 8

````
import numpy as np
from scipy import stats

marks = [6,7,5,7,7,8,7,6,9,7,4,10,6,8,8,9,5,6,4,8]
print ("Mean:", round(np.mean(marks), 2))
print ("Median:", round(np.median(marks), 2))
print ("Mode:", stats.mode(marks))
print ("Standard Deviation:", round(np.std(marks), 2))
````

Q.3. The number of times I go to the gym in weekdays, are given below along with its associated probability:
> x = 0, 1, 2, 3, 4, 5


> f(x) = 0.09, 0.15, 0.40, 0.25, 0.10, 0.01




Calculate the mean no. of workouts in a week. Also evaluate the variance involved in it.

````
x = np.array([0,1,2,3,4,5])
fx = np.array([0.09, 0.15, 0.40, 0.25, 0.10, 0.01])
arr =  x * fx
print ("Mean no.of workouts in a week:",round(np.mean(arr), 2))
print ("Variance involved:", round(np.var(arr), 2))
````
